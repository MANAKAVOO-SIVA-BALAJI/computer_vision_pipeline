```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Preprocess the frame (resize, normalize, etc.)
    frame = cv2.resize(frame, (640, 480))
    frame = frame / 255.0  # Normalize to [0, 1]

    cv2.imshow("Video", frame)
    if cv2.waitKey(1) & 0xFF == 27:  # ESC to exit
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Read an image
import cv2
image = cv2.imread('../entrance_camera_angle.jpg')
if image is not None:
    print("Image Properties:")
    print(f"Shape: {image.shape}")
    print(f"Width: {image.shape[1]} pixels")
    print(f"Height: {image.shape[0]} pixels")
    print(f"Channels: {image.shape[2]}")
    resized_image = cv2.resize(image, (640, 480))


# Read a video
video_path = '../entrance_video_10s.mp4'
video = cv2.VideoCapture(video_path)
if video.isOpened():
    print("\nVideo Properties:")
    width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = video.get(cv2.CAP_PROP_FPS)
    frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"Width: {width} pixels")
    print(f"Height: {height} pixels")
    print(f"FPS: {fps}")
    print(f"Frame Count: {frame_count}")

video.release()

```

```python
cv2.imwrite("entrance_resized.jpg", resized_image)
```
